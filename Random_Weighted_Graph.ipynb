```python
# Source: https://cocalc.com/share/public_paths/embed/e8d6818f7c1587f87fd72cb85b3a4cd09db6c5b3
import random;

def random_weighted_graph(n, p, lb, up):
    # This function creates a random weighted graph, given a number of vertices,
    #   probability, and bounds on the weight.
    # Inputs:
    #         n  = Number of vertices
    #         p  = Probability of adding Edges. 0 < p <= 1
    #         lb = Lower bound on edge weight
    #         up = Upper bound on edge weight
    
    # Output: A Weighted graph with random edges and weights of edges.

    # Create random graph of n vertices with p probability of creating an edge
    H = graphs.RandomGNP(n,p)
    
    # Unweighted edges
    edgesH = H.edges()
    
    # Number of edges
    length = len(edgesH)
    
    # Randomly weight edges
    edgesH = [(edgesH[c][0],edgesH[c][1],random.randint(lb,up)) for c in range(0,length)]
    
    # Create weighted graph H.
    H = Graph(edgesH, weighted = True)
    return H
```

```python
# Calls the function with __ vertices, __ probability, and weights __ to __

num_vertices    = 8
probability     = 0.5
lowest_weight   = 1
greatest_weight = 8

G = random_weighted_graph(num_vertices,probability,lowest_weight,greatest_weight)

# Plot the generated graph

G.show(edge_labels=True)

# Display the edges and weights of G

print(len(G.vertices()))
for edge in G.edges():
    print(str(edge[0]) + "," + str(edge[1]) + "," + str(edge[2]))
```

```python
# Find shortest path by weight

path_from = 0
path_to   = 1

print(G.shortest_path(path_from,path_to,by_weight=True))
print("Total cost: " + str(G.shortest_path_length(path_from,path_to,by_weight=True)))
```
